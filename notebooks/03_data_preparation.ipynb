```python
import pandas as pd

#load the cleaned cvs
customers = pd.read_csv("../data/customers_cleaned.csv")
products = pd.read_csv("../data/products_cleaned.csv")
orders = pd.read_csv("../data/orders_cleaned.csv")
stores = pd.read_csv("../data/stores_cleaned.csv")

#merge all tables into one master dataset
df = orders.merge(customers, on="customer_id", how="left")
df = df.merge(products, on="product_id", how="left")
df = df.merge(stores, on="store_id", how="left")

#revenue, cost, profit, margin
df["revenue"] = df["quantity"] * df["price"]
df["cost_total"] = df["quantity"] * df["cost"]
df["profit"] = df["revenue"] -df["cost_total"]
df["margin_pct"] = (df["profit"] / df["revenue"]).round(3)

#time-based features
df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")
df["year"] = df["order_date"].dt.year
df["month"] = df["order_date"].dt.month
df["month_name"] = df["order_date"].dt.month_name()
df["day"] = df["order_date"].dt.day
df["weekday"] = df["order_date"].dt.day_name()
df["quarter"] = df["order_date"].dt.quarter

#customer features
df["customer_full_name"] = df["first_name"] + " " + df["last_name"]
df["age_group"] = pd.cut(
    df["age"],
    bins=[0, 18, 30, 45, 60, 120],
    labels=["Teen", "Young Adult", "Adult", "Middle Age", "Senior"]
)

#product features
df["is_low_stock"] = df["stock_level"] < 10
df["is_high_margin"] = df["margin_pct"] > 0.30

#store features
df["is_high_footfall"] = df["footfall"] > df["footfall"].median()

#Final output
print("Final dataset shape:", df.shape)
df.to_csv("../data/final_dataset.csv", index=False)
print("Final dataset saved")
df.head()
```
